```python
import pandas as pd

tamilNadu2017 = pd.read_csv("/content/drive/MyDrive/Wind Speed/Madurai, Tamil Nadu Data/3227257_9.93_78.14_2017.csv")
tamilNadu2018 = pd.read_csv("/content/drive/MyDrive/Wind Speed/Madurai, Tamil Nadu Data/3227257_9.93_78.14_2018.csv")
tamilNadu2019 = pd.read_csv("/content/drive/MyDrive/Wind Speed/Madurai, Tamil Nadu Data/3227257_9.93_78.14_2019.csv")
```

```python
frames = [tamilNadu2017, tamilNadu2018, tamilNadu2019]
tamilNaduData = pd.concat(frames)
```

```python
y = tamilNaduData["Wind Speed"]
x = tamilNaduData.drop(['Wind Speed'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=123)
```
